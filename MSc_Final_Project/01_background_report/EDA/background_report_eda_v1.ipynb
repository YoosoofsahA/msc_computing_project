```python
from docx import Document
from docx.shared import Inches

null_counts_df.index.name = 'Patient'

# Create a new Word document
doc = Document()

# Add a title
doc.add_heading('Null Counts Analysis', level=1)

# Function to add a DataFrame to a Word document
def add_dataframe_to_word(doc, df, table_title=None):
    if table_title:
        doc.add_heading(table_title, level=2)

    table = doc.add_table(rows=1, cols=len(df.columns) + 1)
    table.style = 'Table Grid'

    # Add the header row
    hdr_cells = table.rows[0].cells
    hdr_cells[0].text = df.index.name
    for i, column in enumerate(df.columns):
        hdr_cells[i + 1].text = column

    # Add the data rows
    for index, row in df.iterrows():
        row_cells = table.add_row().cells
        row_cells[0].text = str(index)
        for i, cell in enumerate(row):
            row_cells[i + 1].text = str(cell)

    # Fit table to page width (A4 page width is 8.3 inches)
    for row in table.rows:
        for cell in row.cells:
            cell.width = Inches(8.3 / (len(df.columns) + 1))

# Split the DataFrame into chunks if necessary to fit the width of an A4 page
max_cols_per_table = 8  # Adjust this number based on desired fit
num_chunks = (len(null_counts_df.columns) - 1) // max_cols_per_table + 1

for i in range(num_chunks):
    start_col = i * max_cols_per_table
    end_col = start_col + max_cols_per_table
    df_chunk = null_counts_df.iloc[:, start_col:end_col]
    add_dataframe_to_word(doc, df_chunk, f'Table {i + 1}')

# Save the document
output_file = 'content/drive/MyDrive/MSc_Final_Project/Background_Report/EDA/null_counts_analysis.docx'
doc.save(output_file)

print(f"Null counts analysis table has been exported to {output_file}")
```

2.4 Analyse the time taken for intubation for each patient

```python
# Function to calculate the number of rows between the first and last filled value of PsSet
def calculate_rows_between_first_and_last_psset(patient_data):
    psset_series = patient_data['PsSet']

    # Find the first and last non-NaN index in PsSet
    first_index = psset_series.first_valid_index()
    last_index = psset_series.last_valid_index()

    # Calculate the number of rows between the first and last filled row
    number_of_rows = last_index - first_index

    return number_of_rows

# Function to convert minutes to days, hours, and minutes
def convert_minutes(minutes):
    days = minutes // (24 * 60)
    minutes %= (24 * 60)
    hours = minutes // 60
    minutes %= 60
    return days, hours, minutes


# Calculate the number of rows for each patient dataset
rows_1 = calculate_rows_between_first_and_last_psset(patient_data_1)
rows_2 = calculate_rows_between_first_and_last_psset(patient_data_2)
rows_3 = calculate_rows_between_first_and_last_psset(patient_data_3)
rows_4 = calculate_rows_between_first_and_last_psset(patient_data_4)
rows_5 = calculate_rows_between_first_and_last_psset(patient_data_5)

print(f"patient_data_1: {rows_1}")
print(f"patient_data_2: {rows_2}")
print(f"patient_data_3: {rows_3}")
print(f"patient_data_4: {rows_4}")
print(f"patient_data_5: {rows_5}")

# Convert to days and hours
days_1, hours_1, mins_1 = convert_minutes(rows_1)
days_2, hours_2, mins_2 = convert_minutes(rows_2)
days_3, hours_3, mins_3 = convert_minutes(rows_3)
days_4, hours_4, mins_4 = convert_minutes(rows_4)
days_5, hours_5, mins_5 = convert_minutes(rows_5)

print(f"Patient 1: {rows_1} minutes is equivalent to {days_1} days, {hours_1} hours, and {mins_1} minutes.")
print(f"Patient 2: {rows_2} minutes is equivalent to {days_2} days, {hours_2} hours, and {mins_2} minutes.")
print(f"Patient 3: {rows_3} minutes is equivalent to {days_3} days, {hours_3} hours, and {mins_3} minutes.")
print(f"Patient 4: {rows_4} minutes is equivalent to {days_4} days, {hours_4} hours, and {mins_4} minutes.")
print(f"Patient 5: {rows_5} minutes is equivalent to {days_5} days, {hours_5} hours, and {mins_5} minutes.")


```
